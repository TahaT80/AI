# kmeans

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris , make_blobs

from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score
```

```python
x, y = make_blobs(n_samples=300, n_features=2, centers=3)
x.shape, y
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
center = model.cluster_centers_
labels = model.labels_
center,labels
```

```python
plt.scatter(x[:,0],x[:,1],c=labels)  #******************c
plt.scatter(center[:,0],center[:,1],c='red',marker='x')
plt.show()
```

```python
x,y = load_iris(return_X_y=True)
x.shape,y.shape
```

```python
plt.scatter(x[:,0],x[:,1], marker='.')
```

```python
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
model.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,3], marker='.')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,3],c='r', marker='o')
```

```python
p =model.predict(x)
p
```

```python
plt.scatter(x[:,0],x[:,3], marker='.',c=p)
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,3],c='r', marker='o')
```

```python
inertias = []
for i in range(2,20):
    model = KMeans(n_clusters=i)
    model.fit(x)
    inertias.append(model.inertia_)  # ÙØ§ØµÙ„Ù‡ Ø¹Ø¯Ø¯ Ù‡Ø§
```

```python
plt.plot(np.arange(2,20),inertias)
```
