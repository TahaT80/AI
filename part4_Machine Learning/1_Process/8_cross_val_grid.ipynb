```python
%matplotlib inline
import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score ,cross_val_predict 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV  # مثل بالاییه ولی سریع تر و دقت کمتری نسبت به بالایی داره
```

```python
from sklearn.datasets import load_iris
x,y = load_iris(return_X_y=True)
x.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
X_train.shape,X_test.shape
```

```python
model = KNeighborsClassifier(n_neighbors=2)
```

# K-Fold

### cross_val
- میخاد بگه مدل خوبه یا نه و ما باید چند تا مدل داشته که ببینیم کدوم مدل بهتره
- برای پیدا کردن بهترین هاپر پارامتر 

```python
cross_val_score(model,X_train,y_train,cv=5)  # k-fold 5 
```

```python
cross_val_predict(model,X_test,y_test,cv=5)  
```

```python
for i in range(3,23,2):
    model = KNeighborsClassifier(n_neighbors=i)
    # print(cross_val_score(model,X_train,y_train,cv=5))
    print(f'{i} : {np.mean(cross_val_score(model,X_train,y_train,cv=5))}')
```

```python
model = KNeighborsClassifier()

params = {"weights": ["uniform", "distance"], "n_neighbors": list(range(3, 10))}

grid = GridSearchCV(model,params)
grid.fit(X_train,y_train)
```

```python
grid.cv_results_
```

```python
grid.best_score_
```
