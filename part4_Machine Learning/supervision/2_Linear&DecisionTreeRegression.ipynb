```python
%matplotlib inline
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4]).reshape(-1,1)
y = np.array([1,3,5,7])
x
```

```python
plt.plot(x,y,'*')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
m = model.coef_[0]
c = model.intercept_
print(m,c)
```

```python
plt.plot(x,y,'*')
plt.plot(x,model.predict(x))
plt.show()
```

```python
result = model.predict([[5]])
result
```

```python
years_of_ex = np.array([0,1,2,3,4,6,8,10]).reshape(-1,1)
salaries = np.array([3000,4000,5500,6000,7000,8000,8500,8700])
plt.plot(years_of_ex,salaries,'*')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(max_depth=3)

model.fit(years_of_ex,salaries)
plt.plot(years_of_ex,salaries,'*')
plt.plot(years_of_ex,model.predict(years_of_ex))
plt.show()
```
