````python
import pandas as pd
import numpy as np
````

````python

df1 = pd.DataFrame({
    'id':[1,2,3,4,5,6],
    'name': ['ali','mmd','taha','adadw','wadammd','wda'],
})
df2 = pd.DataFrame({
    'id':[1,2,3,4,5,6],
    'Salary':['20','10','43','33','66','12']
})
df = pd.merge(df1,df2,on="id")
df
````

````python
df = pd.DataFrame({
    "ID": [1, 2, 2, 3, 4, 5],
    "Name": ["Ali", "Sara", "Sara", "Reza", "Neda", None],
    "Age": [25, 30, np.nan, 28, np.nan, 35]
})
df
````

````python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df = df.drop_duplicates('Name')
````

````python
dates = pd.date_range("2025-01-01", "2025-01-31")
import numpy as np
df = pd.DataFrame({"Date": dates, "Sales": np.random.randint(100, 500, len(dates))})
df.set_index("Date", inplace=True)


# وقتی df یه ستون تاریخ (Date) به عنوان index داشته باشه، می‌تونیم داده‌ها رو بر اساس بازه زمانی گروه‌بندی کنیم.

# "W" یعنی هفته‌ای (Weekly).

# sum() میگه: مقادیر هر هفته رو جمع کن.```


# مجموع فروش هفتگی
weekly = df.resample("W").sum()  
print("Weekly Sales:\n", weekly)

# میانگین فروش ماهانه
monthly = df.resample("M").mean()
print("\nMonthly Avg Sales:\n", monthly)
df
````

````python
df["MA7"] = df["Sales"].rolling(window=7).mean()
print(df.head(10))
````

````python
df = pd.DataFrame({
    "Name":["Ali","Sara","Sara","Reza","Neda", None],
    "Age":[25,30,np.nan,28,np.nan,35],
    "Salary":[5000,np.nan,6000,7000,6500,np.nan],
    "Dept":["IT","HR","HR","IT",None,"Sales"]
})
df
````

````python

# پر کردن Salary با میانگین دپارتمان خودش
df["Salary"] = df.groupby("Dept")["Salary"].transform(lambda x: x.fillna(x.mean()))
df
````

````python
df.groupby("Dept").value_counts()
````

````python
df.groupby("Dept")["Salary"].value_counts()
````

````python
result = (df.query("Salary > 5000 and Dept == 'IT'")
            [["Name","Salary"]])
print(result)
````

````python
df["Dept"] = df["Dept"].astype("category")
print(df["Dept"].value_counts())
````
