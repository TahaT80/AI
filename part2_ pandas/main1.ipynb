# changing data in python list 

```python
a = ['Y','N','12',12,'test']
a_prime = []
for item in a:
    if item not in ['Y','N']:
        a_prime.append('NAN')
    else: 
        a_prime.append(item)
a_prime
```

```python
a = ['Y','N','12',12,'test']
a_prime = ['NAN' if item not in ['Y','N'] else item for item in a ]
a_prime
```

```python
def changer(item):
    if item not in ['Y','N']:
        return 'NAN'
    return item

a_prime = list(map(changer,a))
a_prime
```

```python
a_prime = list(map(lambda item:'NAN' if item not in ['Y','N'] else item, a ))
a_prime
```
