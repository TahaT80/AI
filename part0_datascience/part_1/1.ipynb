# regex

```python
regex = re.compile(r'<strong[^>]*(.*?)<\/strong>')



r = regex.search(content)

r.group(1)

# <strong class="yekanbakh-en left-8 line-clamp-3 text-lg font-semiBold leading-8.5 -tracking-0.5 text-primary-shade-1 xl:line-clamp-2" id="pdp_name">گوشی موبايل سامسونگ Galaxy A36 5G ظرفیت 256 گیگابایت رم 8 گیگابایت - ویتنام</strong></h1></div>
```

# BeautifulSoup

```python
dom  = BeautifulSoup(content,'html.parser')

items = dom.select('#pdp_name') # css selector

print(len(items))

item = items[0]
item.text
```
