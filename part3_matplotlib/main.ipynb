```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 3, 4, 6, 3, 2, 6, 7, 8]

# plt.plot(x, y,'*', color="green")
plt.plot(x, y,marker='*', color="green")
plt.title("plot")
plt.xlabel('tahaX')
plt.ylabel('tahay')
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [4, 6, 2, 8]
plt.bar(x, y, color="red")
plt.title("bar")
plt.xlabel("color")
plt.ylabel("count")
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [4, 6, 2, 8]
plt.pie(y, labels=x, autopct="%1.1f%%", colors=["red", "green", "blue", "orange"])
plt.title("example")

plt.show()
```

```python
import numpy as np

x = np.random.rand(1000,1)

plt.hist(x,bins=5)
plt.show()
```
