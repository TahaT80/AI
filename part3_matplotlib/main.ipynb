```python
x= np.linspace(1,360,100)
y = np.sin(np.radians(x))
y_2 = np.log(x)


plt.plot(x,y,'r')

plt.plot(x,y_2,'b')
```

```python
x= np.linspace(1,360,100)
y = np.sin(np.radians(x))
y_2 = np.log(x)

fig,ax = plt.subplots()


ax.plot(x,y,'r')
ax.set_ylabel('$sin(x)$')
ax2 = ax.twinx() # کپی از نمودار با x مشترک
ax2.plot(x,y_2,'b')
ax2.set_yticks(y_2,minor=True)
ax2.set_ylabel('$log(x)$')
```
