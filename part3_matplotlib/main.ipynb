```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 3, 4, 6, 3, 2, 6, 7, 8]

# plt.plot(x, y,'*', color="green")
plt.plot(x, y,marker='*', color="green")
plt.title("plot")
plt.xlabel('tahaX')
plt.ylabel('tahay')
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [4, 6, 2, 8]
plt.bar(x, y, color="red")
plt.title("bar")
plt.xlabel("color")
plt.ylabel("count")
plt.show()
```

```python
x = ["red", "green", "blue", "orange"]
y = [4, 6, 2, 8]
plt.pie(y, labels=x, autopct="%1.1f%%", colors=["red", "green", "blue", "orange"])
plt.title("example")

plt.show()
```

```python
x= np.linspace(0,360,50)
y = np.sin(np.radians(x))
y_2 = np.cos(np.radians(x))
```

```python
plt.plot(x,y)
```

```python
plt.plot(y)   # مقدار تولید میشه با کمک
            # Arange
```

```python
plt.plot(x,y,'o-r')
```

```python
plt.plot(x,y)
plt.plot(x,y_2)
plt.savefig('out.png')
```

```python
plt.plot(x,y)
plt.plot(x,y_2)
plt.legend(['$Sin(x)$','Cos'])
plt.title('some plots')
plt.xlabel('values')
plt.ylabel('$Sin & Cos$')
plt.grid(which='both')
plt.xlim(0,360)
plt.ylim(-1,1)
plt.axis('on');
```
